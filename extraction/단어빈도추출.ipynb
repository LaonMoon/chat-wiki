```python
import pandas as pd

# 데이터프레임을 생성합니다.
df = pd.read_excel('/content/drive/MyDrive/result.xlsx')

df_grouped = df.groupby('thld_20min').agg({'chat': ' '.join})
```

```python
df_grouped
```

```python
from collections import Counter
from konlpy.tag import Okt

okt = Okt()

word_counts = []

for index, row in df_grouped.iterrows():
    text = row['chat']
    nouns = okt.nouns(text)
    count = Counter(nouns)
    word_counts.append(count.most_common(5))

word_counts
```

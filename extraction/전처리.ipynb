```python
import pandas as pd

# 데이터프레임을 생성합니다.
df = pd.read_excel('/content/drive/MyDrive/result.xlsx')

# 'column_name' 열에서 'keyword'를 포함하는 행 추출하여 새로운 데이터프레임에 저장
extracted_df = df[df['chat'].str.contains('https')]

# 추출된 데이터를 제외한 나머지 데이터 선택하여 새로운 데이터프레임 생성
df = df.drop(extracted_df.index)

# 'column_name' 열에서 'keyword'를 포함하는 행 제거
df = df[~df['chat'].str.contains('안녕|ㅋㅋㅋㅋㅋ|사진|이모티콘|아프지|회원에게|삭제된|감사합|네')]

df['chat'] = df['chat'].str.replace('[,|;|^|ㅎ|"|ㄷ|ㅜ]', '')

```

```python
df_grouped = df.groupby('thld_20min').agg({'chat': ' '.join})
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt

okt = Okt()
word_counts = []

for index, row in df_grouped.iterrows():
    text = row['chat']
    nouns = okt.nouns(text)
    preprocessed_text = ' '.join(nouns)

    # TF-IDF 계산
    vectorizer = TfidfVectorizer()
    tf_idf = TfidfVectorizer().fit([preprocessed_text])
    sorted_items = sorted(tf_idf.vocabulary_.items(), key=lambda x: x[1], reverse=True)[:5]
    word_counts.append(sorted_items)

word_counts
```
